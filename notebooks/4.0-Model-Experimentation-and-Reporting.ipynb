```python
# notebooks/4.0-Model-Experimentation-and-Reporting.ipynb

# 添加项目根目录到Python路径
import sys
import os
# 获取当前文件所在的目录路径
current_dir = os.path.dirname(os.path.abspath('__file__'))
# 获取项目根目录路径
project_root = os.path.abspath(os.path.join(current_dir, '..'))
# 添加项目根目录到Python路径
sys.path.append(project_root)
print(f"项目根目录已添加到Python路径: {project_root}")

# 导入所需库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_curve, auc
from sklearn.inspection import permutation_importance # 用于特征重要性
from sklearn.linear_model import LogisticRegression  # 添加LogisticRegression的导入

# 导入自定义模块
from src.data_loader import load_data
from src.preprocessor import preprocess_data
from src.models import train_model, evaluate_model, save_model, load_model
from src.visualizer import ( # 导入所有需要的可视化函数
    plot_churn_distribution,
    plot_correlation_matrix,
    plot_numerical_distributions,
    plot_categorical_churn_rates,
    plot_roc_curves,
    plot_feature_importance,
    plot_confusion_matrix
)
from src.config import PROCESSED_DATA_PATH, TARGET_COLUMN, MODEL_PATH, RANDOM_STATE, FIGURES_DIR, REPORTS_DIR

# 设置图表风格
sns.set_style("whitegrid")
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS'] # 尝试多种字体支持中文
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
# plt.tight_layout() # 在全局设置，但在每个图表函数内部也会调用以确保布局

# 确保 FIGURES_DIR 和 REPORTS_DIR 存在
import os
os.makedirs(FIGURES_DIR, exist_ok=True)
os.makedirs(REPORTS_DIR, exist_ok=True)


print("--- 开始模型实验与报告生成 ---")

# 1. 加载并预处理数据
print("加载原始数据并进行预处理...")
raw_df = load_data()
df_processed = preprocess_data(raw_df)

if df_processed is None:
    print("数据加载或预处理失败，无法进行模型实验。")
    exit()

X = df_processed.drop(TARGET_COLUMN, axis=1)
y = df_processed[TARGET_COLUMN]

# --- 再次进行一些关键的EDA可视化 (使用visualizer) ---
print("\n--- 重新生成关键EDA可视化图表 ---")

# 流失分布图
plot_churn_distribution(df_processed,
                        save_path=os.path.join(FIGURES_DIR, 'churn_distribution.png'))

# 相关性矩阵图
plot_correlation_matrix(df_processed.drop(TARGET_COLUMN, axis=1), # 只对特征进行相关性分析，不包括目标变量自身
                        save_path=os.path.join(FIGURES_DIR, 'feature_correlation_matrix.png')) # 更明确的文件名

print(f"\n特征与目标变量 '{TARGET_COLUMN}' 的相关性:")
correlation_matrix_with_target = df_processed.corr(numeric_only=True)
churn_correlations = correlation_matrix_with_target[TARGET_COLUMN].sort_values(ascending=False)
display(churn_correlations.drop(TARGET_COLUMN)) # 排除自身与自身的完美相关

# 数值特征与流失的关系图
numerical_features_to_plot = ['tenure', 'MonthlyCharges', 'TotalCharges']
plot_numerical_distributions(df_processed, numerical_features_to_plot,
                             hue_column=TARGET_COLUMN,
                             title_suffix=' (与流失相关)',
                             save_path=os.path.join(FIGURES_DIR, 'numerical_features_vs_churn.png'))

# 分类特征与流失率的关系图
selected_categorical_for_churn_viz = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
    'InternetService_Fiber optic', 'Contract_Month-to-month',
    'PaperlessBilling', 'PaymentMethod_Electronic check',
    'OnlineSecurity_No', 'TechSupport_No'
]
# 确保这些列在绘图前是int类型，为了plot_categorical_churn_rates内部统一处理
for col in selected_categorical_for_churn_viz:
    if col in df_processed.columns and df_processed[col].dtype == 'bool':
        df_processed[col] = df_processed[col].astype(int)

plot_categorical_churn_rates(df_processed, selected_categorical_for_churn_viz,
                             save_path=os.path.join(FIGURES_DIR, 'categorical_churn_rates.png'))


# 存储所有模型的评估结果、模型对象和测试集预测，以便后续可视化和报告
model_results = {}
models_for_roc = {} # {model_name: (model_obj, X_test, y_test)}
model_test_predictions = {} # {model_name: (y_test, y_test_pred)}


# --- 模型 1: 逻辑回归 (Logistic Regression) ---
print("\n##### 训练和评估 Logistic Regression 模型 #####")
lr_model, lr_train_pred, lr_test_pred, y_test_lr, X_test_lr = train_model(X, y, model_name='LogisticRegression', use_smote=False)
lr_metrics = evaluate_model(lr_model, X_test_lr, y_test_lr, lr_train_pred, lr_test_pred, model_name='LogisticRegression')
model_results['Logistic Regression'] = lr_metrics
models_for_roc['Logistic Regression'] = (lr_model, X_test_lr, y_test_lr)
model_test_predictions['Logistic Regression'] = (y_test_lr, lr_test_pred)
save_model(lr_model, MODEL_PATH.replace('.pkl', '_lr.pkl'))


# --- 模型 2: 随机森林 (Random Forest) (不使用 SMOTE) ---
print("\n##### 训练和评估 Random Forest 模型 (无 SMOTE) #####")
rf_model, rf_train_pred, rf_test_pred, y_test_rf, X_test_rf = train_model(X, y, model_name='RandomForest', use_smote=False, n_estimators=100, max_depth=10)
rf_metrics = evaluate_model(rf_model, X_test_rf, y_test_rf, rf_train_pred, rf_test_pred, model_name='Random Forest (No SMOTE)')
model_results['Random Forest (No SMOTE)'] = rf_metrics
models_for_roc['Random Forest (No SMOTE)'] = (rf_model, X_test_rf, y_test_rf)
model_test_predictions['Random Forest (No SMOTE)'] = (y_test_rf, rf_test_pred)
save_model(rf_model, MODEL_PATH.replace('.pkl', '_rf_no_smote.pkl'))


# --- 模型 3: 随机森林 (Random Forest) (使用 SMOTE) ---
print("\n##### 训练和评估 Random Forest 模型 (带 SMOTE) #####")
rf_smote_model, rf_smote_train_pred, rf_smote_test_pred, y_test_rf_smote, X_test_rf_smote = \
    train_model(X, y, model_name='RandomForest', use_smote=True, n_estimators=100, max_depth=10)
rf_smote_metrics = evaluate_model(rf_smote_model, X_test_rf_smote, y_test_rf_smote, rf_smote_train_pred, rf_smote_test_pred, model_name='Random Forest (SMOTE)')
model_results['Random Forest (SMOTE)'] = rf_smote_metrics
models_for_roc['Random Forest (SMOTE)'] = (rf_smote_model, X_test_rf_smote, y_test_rf_smote)
model_test_predictions['Random Forest (SMOTE)'] = (y_test_rf_smote, rf_smote_test_pred)
save_model(rf_smote_model, MODEL_PATH.replace('.pkl', '_rf_smote.pkl'))


# --- 模型 4: 梯度提升 (Gradient Boosting) (带 SMOTE) ---
print("\n##### 训练和评估 Gradient Boosting 模型 (带 SMOTE) #####")
gb_model, gb_train_pred, gb_test_pred, y_test_gb, X_test_gb = \
    train_model(X, y, model_name='GradientBoosting', use_smote=True, n_estimators=100, learning_rate=0.1, max_depth=3)
gb_metrics = evaluate_model(gb_model, X_test_gb, y_test_gb, gb_train_pred, gb_test_pred, model_name='Gradient Boosting (SMOTE)')
model_results['Gradient Boosting (SMOTE)'] = gb_metrics
models_for_roc['Gradient Boosting (SMOTE)'] = (gb_model, X_test_gb, y_test_gb)
model_test_predictions['Gradient Boosting (SMOTE)'] = (y_test_gb, gb_test_pred)
save_model(gb_model, MODEL_PATH.replace('.pkl', '_gb_smote.pkl'))


# --- 可选：使用 GridSearchCV 进行超参数调优 (以逻辑回归为例) ---
print("\n##### 使用 GridSearchCV 进行超参数调优 (Logistic Regression) #####")
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}
# 为了快速演示，这里使用较小的数据集划分，实际应用中应该使用完整数据集
X_tune, _, y_tune, _ = train_test_split(X, y, test_size=0.8, random_state=RANDOM_STATE, stratify=y) # 仅用20%数据快速演示

grid_search_lr = GridSearchCV(LogisticRegression(random_state=RANDOM_STATE, solver='liblinear'), param_grid_lr, cv=5, scoring='roc_auc', n_jobs=-1, verbose=0) # verbose=0 减少输出
grid_search_lr.fit(X_tune, y_tune)

print(f"最佳超参数: {grid_search_lr.best_params_}")
print(f"最佳 ROC AUC Score: {grid_search_lr.best_score_:.4f}")

# --- 模型比较 ---
print("\n##### 所有模型性能比较 #####")
results_df = pd.DataFrame(model_results).T
display(results_df[['accuracy', 'precision', 'recall', 'f1_score', 'ROC_AUC']].sort_values(by='ROC_AUC', ascending=False))

# --- 绘制 ROC 曲线 (使用 visualizer) ---
print("\n##### 绘制 ROC 曲线 #####")
plot_roc_curves(models_for_roc, save_path=os.path.join(FIGURES_DIR, 'model_roc_curves.png'))


# --- 特征重要性分析 (以最好的模型为例，基于 ROC AUC) ---
print("\n##### 特征重要性分析 #####")
best_model_name_for_fi = results_df['ROC_AUC'].idxmax()
best_model_obj, X_test_for_fi, y_test_for_fi = models_for_roc[best_model_name_for_fi]

# 检查模型是否支持feature_importances_属性 (如树模型)
# 如果是逻辑回归，我们用permutation importance
if hasattr(best_model_obj, 'feature_importances_'):
    print(f"使用模型 {best_model_name_for_fi} 的 feature_importances_ 属性。")
    feature_importances_raw = best_model_obj.feature_importances_
    feature_names = X.columns
    feature_importances_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importances_raw
    }).sort_values(by='Importance', ascending=False)
else:
    print(f"使用 Permutation Importance 分析模型 {best_model_name_for_fi} 的特征重要性。")
    # 使用 permutation importance 更可靠，因为它考虑了模型在数据上的实际表现
    result = permutation_importance(best_model_obj, X_test_for_fi, y_test_for_fi, n_repeats=10, random_state=RANDOM_STATE, n_jobs=-1)
    sorted_idx = result.importances_mean.argsort()[::-1] # 降序排序
    feature_importances_df = pd.DataFrame({
        'Feature': X_test_for_fi.columns[sorted_idx],
        'Importance': result.importances_mean[sorted_idx]
    })

print("Top 15 特征重要性:")
display(feature_importances_df.head(15))
plot_feature_importance(feature_importances_df, top_n=15,
                        title=f'{best_model_name_for_fi} 特征重要性',
                        save_path=os.path.join(FIGURES_DIR, 'feature_importance.png'))


# --- 绘制最佳模型的混淆矩阵 ---
best_model_name_for_cm = results_df['ROC_AUC'].idxmax()
y_test_best, y_pred_best = model_test_predictions[best_model_name_for_cm]
print(f"\n##### 绘制最佳模型 '{best_model_name_for_cm}' 的混淆矩阵 #####")
plot_confusion_matrix(y_test_best, y_pred_best,
                      title=f'{best_model_name_for_cm} 混淆矩阵',
                      save_path=os.path.join(FIGURES_DIR, 'best_model_confusion_matrix.png'))


# --- 生成最终报告 ---
print("\n##### 生成最终报告 #####")
report_path = os.path.join(REPORTS_DIR, 'churn_analysis_report.md')

with open(report_path, 'w', encoding='utf-8') as f:
    f.write("# 电信客户流失数据分析及可视化报告\n\n")
    f.write("## 1. 项目概述\n")
    f.write("本项目旨在对电信客户流失数据进行深入分析，识别流失客户的特征，预测流失趋势，并可视化呈现分析结果，以帮助电信公司制定有效的客户保留策略。\n\n")
    f.write("## 2. 数据集描述与预处理\n")
    f.write(f"数据集包含 {df_processed.shape[0]} 条客户记录和 {df_processed.shape[1]} 个特征 (预处理后)。\n")
    f.write("主要处理步骤包括：缺失值填充 (TotalCharges)、客户ID删除、二元特征标签编码、多元分类特征独热编码、数值特征标准化。\n\n")
    f.write("## 3. 探索性数据分析 (EDA) 关键发现\n")
    f.write("通过EDA，我们发现：\n")
    churn_no_percentage = (df_processed[TARGET_COLUMN].value_counts(normalize=True)[0] * 100) if 0 in df_processed[TARGET_COLUMN].value_counts(normalize=True) else 0
    churn_yes_percentage = (df_processed[TARGET_COLUMN].value_counts(normalize=True)[1] * 100) if 1 in df_processed[TARGET_COLUMN].value_counts(normalize=True) else 0
    f.write(f"- 目标变量 'Churn' (客户流失) 存在类别不平衡现象：不流失客户占 {churn_no_percentage:.2f}%，流失客户占 {churn_yes_percentage:.2f}%。\n")
    f.write("- **流失风险最高的客户特征：**\n")
    f.write("    - **合同类型：** 按月付费 (Month-to-month) 合同的客户流失率远高于一年或两年合同的客户。\n")
    f.write("    - **服务类型：** 未订阅在线安全、技术支持服务，以及使用光纤互联网服务的客户流失率较高。\n")
    f.write("    - **支付方式：** 使用电子支票 (Electronic check) 支付的客户流失率较高。\n")
    f.write("    - **客户时长 (Tenure)：** 新客户（Tenure较短）的流失率显著更高。\n")
    f.write("    - **月费用 (MonthlyCharges)：** 月费用较高的客户流失率可能也更高。\n\n")
    f.write("![客户流失分布](figures/churn_distribution.png)\n\n")
    f.write("![特征相关性矩阵](figures/feature_correlation_matrix.png)\n\n")
    f.write("![数值特征与流失关系](figures/numerical_features_vs_churn.png)\n\n")
    f.write("![分类特征流失率](figures/categorical_churn_rates.png)\n\n")


    f.write("## 4. 机器学习模型构建与评估\n")
    f.write("我们训练并评估了多种分类模型，并对比了它们在流失预测上的性能。为解决类别不平衡问题，部分模型使用了SMOTE过采样技术。\n\n")
    f.write("### 模型性能概览\n")
    f.write(results_df[['accuracy', 'precision', 'recall', 'f1_score', 'ROC_AUC']].sort_values(by='ROC_AUC', ascending=False).to_markdown(numalign="left", stralign="left") + "\n\n")
    
    best_model_name = results_df['ROC_AUC'].idxmax()
    best_model_metrics = model_results[best_model_name]
    f.write(f"基于 ROC AUC Score，最佳模型为：**{best_model_name}**，其 ROC AUC 值为 **{best_model_metrics['ROC_AUC']:.4f}**。\n\n")

    f.write("![模型ROC曲线](figures/model_roc_curves.png)\n\n")
    
    f.write(f"### 最佳模型 ({best_model_name}) 详细评估\n")
    f.write(f"- 准确率 (Accuracy): {best_model_metrics['accuracy']:.4f}\n")
    f.write(f"- 精确率 (Precision): {best_model_metrics['precision']:.4f}\n")
    f.write(f"- 召回率 (Recall): {best_model_metrics['recall']:.4f}\n")
    f.write(f"- F1-Score: {best_model_metrics['f1_score']:.4f}\n")
    f.write(f"- ROC AUC Score: {best_model_metrics['ROC_AUC']:.4f}\n\n")
    f.write(f"![{best_model_name} 混淆矩阵](figures/best_model_confusion_matrix.png)\n\n")


    f.write("## 5. 流失原因分析 (特征重要性)\n")
    f.write("通过特征重要性分析 (主要采用Permutation Importance)，我们识别出对客户流失预测影响最大的特征：\n")
    f.write(feature_importances_df.head(10).to_markdown(numalign="left", stralign="left") + "\n\n")
    f.write("这些重要特征再次验证了EDA中的发现，即合同类型、服务附加项和支付方式是影响客户流失的关键因素。\n\n")
    f.write("![特征重要性](figures/feature_importance.png)\n\n")

    f.write("## 6. 结论与建议\n")
    f.write("基于以上分析，我们得出以下结论和客户保留建议：\n")
    f.write("1.  **关注按月合同客户：** 这类客户流失风险最高。可考虑为他们提供长期合同优惠或捆绑服务，增加合同的粘性。\n")
    f.write("2.  **提升增值服务订阅率：** 鼓励客户订阅在线安全和技术支持服务，这些服务能显著降低流失率。\n")
    f.write("3.  **优化光纤互联网和电子支票支付体验：** 尽管这些服务可能具有吸引力，但它们的用户流失率较高。需要深入研究这些服务中是否存在痛点或竞争劣势。这可能包括改进服务质量、提供更稳定的连接或更便捷的支付流程。\n")
    f.write("4.  **重视新客户维系：** 在客户入网初期提供更好的引导、专属服务或新客优惠，建立早期忠诚度。\n")
    f.write("5.  **实施差异化营销：** 根据客户的特征（如合同、服务类型、月费用等），对不同流失风险的客户群体采取定制化的挽留策略。\n\n")
    f.write("## 7. 未来工作\n")
    f.write("1.  **更复杂的模型：** 尝试XGBoost、LightGBM等更强大的Boosting模型，或深度学习模型。\n")
    f.write("2.  **高级特征工程：** 探索创建更多交互特征或多项式特征。\n")
    f.write("3.  **时间序列分析：** 如果有客户行为的时间序列数据，可以进行更精细的流失趋势预测和动态风险评估。\n")
    f.write("4.  **A/B 测试：** 将模型预测结果应用于实际营销活动，并进行A/B测试验证挽留策略的有效性。\n")
    f.write("5.  **交互式仪表板：** 使用Streamlit或Dash构建交互式应用，方便业务用户探索数据和预测结果。\n")

print(f"最终报告已生成到 '{report_path}'。")

print("\n--- 项目完成 ---")
```

模型性能比较：
Logistic Regression 表现相当不错，在 ROC AUC 上略微领先 (0.841975)。
Gradient Boosting (SMOTE) 和 Random Forest (SMOTE) 也表现良好，ROC AUC 分别为 0.840487 和 0.837814。
引入 SMOTE 后，随机森林的 Recall 值有所提高 (从 0.5214 到 0.7032)，这表明模型在识别流失客户方面变得更强，但在 Precision 上有所下降 (从 0.6478 到 0.5560)。这符合SMOTE的预期效果，即以一定的精确率损失换取召回率的提升，这在流失预测这种重视召回率的场景中通常是可接受的。
超参数调优： 逻辑回归的最佳超参数为 C=1, penalty='l1'，并且最佳 ROC AUC Score 为 0.8482，略高于初始逻辑回归模型，这说明调优是有效的。
特征重要性：
Contract_Month-to-month (按月合同) 是最重要的流失驱动因素，其重要性显著高于其他特征。
OnlineSecurity_Yes/No (在线安全服务) 和 TechSupport_Yes/No (技术支持服务) 也非常重要。
InternetService_DSL 和 InternetService_Fiber optic（光纤）作为互联网服务类型也对流失有重要影响。
PaymentMethod_Electronic check (电子支票支付) 也是一个重要的因素。
