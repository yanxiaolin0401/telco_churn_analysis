```python
# notebooks/2.0-EDA-Further-Analysis.ipynb

# 添加项目根目录到Python路径
import sys
import os
# 获取当前文件所在的目录路径
current_dir = os.path.dirname(os.path.abspath('__file__'))
# 获取项目根目录路径
project_root = os.path.abspath(os.path.join(current_dir, '..'))
# 添加项目根目录到Python路径
sys.path.append(project_root)
print(f"项目根目录已添加到Python路径: {project_root}")

# 导入所需库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# 导入自定义模块
from src.data_loader import load_data
from src.config import PROCESSED_DATA_PATH, TARGET_COLUMN, FIGURES_DIR # 使用PROCESSED_DATA_PATH

# 设置图表风格
sns.set_style("whitegrid")
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS'] # 尝试多种字体支持中文
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
plt.tight_layout() # 自动调整布局，防止标签重叠

# 确保 FIGURES_DIR 存在
import os
os.makedirs(FIGURES_DIR, exist_ok=True)


print("开始进行更深入的数据探索...")

# 1. 加载预处理后的数据
df_processed = load_data(PROCESSED_DATA_PATH)
# 验证数据结构和类型，确认预处理效果。
if df_processed is not None:
    print("\n--- 预处理后数据概览 ---")
    print(f"数据集形状 (行, 列): {df_processed.shape}")
    print("\n数据集前5行:")
    display(df_processed.head())

    print("\n数据集基本信息:")
    df_processed.info() # 显示数据类型和非空计数

    # 重新确认目标变量分布
    print(f"\n--- 目标变量 '{TARGET_COLUMN}' 分布 (编码后) ---")
    churn_counts = df_processed[TARGET_COLUMN].value_counts() #确认 Churn 列（现在是0/1编码）的分布。
    churn_percentages = df_processed[TARGET_COLUMN].value_counts(normalize=True) * 100
    churn_distribution = pd.DataFrame({'Count': churn_counts, 'Percentage (%)': churn_percentages})
    display(churn_distribution) # 显示目标变量分布

    # 2. 相关性分析
    print("\n--- 相关性分析 ---")
    # 针对数值特征和目标变量计算相关性矩阵
    # 注意：这里会包含标准化后的二元特征，但它们仍然可以进行相关性分析
    correlation_matrix = df_processed.corr(numeric_only=True)

    plt.figure(figsize=(18, 15))
    #使用Seaborn的 heatmap 可视化相关性矩阵，帮助我们识别强相关和弱相关的特征。
    sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('特征相关性矩阵', fontsize=18)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.savefig(os.path.join(FIGURES_DIR, 'correlation_matrix.png'))
    plt.show()
    #特别关注与 TARGET_COLUMN 的相关性，找出与客户流失最相关的特征。
    print(f"\n特征与目标变量 '{TARGET_COLUMN}' 的相关性:")
    churn_correlations = correlation_matrix[TARGET_COLUMN].sort_values(ascending=False) # 按相关性排序
    display(churn_correlations.drop(TARGET_COLUMN)) # 排除自身与自身的完美相关，显示其他特征与流失的相关性

    # 3. 单变量分析 (关键数值特征分布)
    print("\n--- 数值特征分布分析 ---")
    #使用 histplot 绘制 tenure, MonthlyCharges, TotalCharges 的分布图，可以观察它们在标准化后的形态。
    numerical_features = ['tenure', 'MonthlyCharges', 'TotalCharges']
    
    plt.figure(figsize=(15, 5))
    for i, feature in enumerate(numerical_features):
        plt.subplot(1, len(numerical_features), i + 1)
        sns.histplot(df_processed[feature], kde=True, bins=30, color='skyblue')
        plt.title(f'{feature} 分布')
        plt.xlabel(feature)
        plt.ylabel('客户数量')
    plt.tight_layout()
    plt.savefig(os.path.join(FIGURES_DIR, 'numerical_feature_distributions.png'))
    plt.show()

    # 4. 双变量分析 (特征与流失率的关系)
    print("\n--- 特征与流失率的关系分析 ---")

    # 将布尔类型的独热编码列转换为数值类型（如果它们不是自动转换为0/1）
    # 确保所有独热编码列都是0或1
    for col in df_processed.select_dtypes(include='bool').columns:
        df_processed[col] = df_processed[col].astype(int) # 将布尔类型转换为整数类型（0和1）

    # 针对分类特征绘制  流失率 条形图
    # 识别原始的分类特征（或其独热编码前身的逻辑分组）
    categorical_original_features = [
        'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
        'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
        'Contract', 'PaperlessBilling', 'PaymentMethod'
    ]

    # 绘制不同类别特征的流失率,选择了一些重要的分类特征（或其独热编码后的代表列），通过分组计算流失率并用条形图展示。
    # 为了简化，我们选择一些重要的进行可视化
    selected_categorical_for_churn_viz = [
        'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
        'InternetService_Fiber optic', 'Contract_Month-to-month',
        'PaperlessBilling', 'PaymentMethod_Electronic check',
        'OnlineSecurity_No', 'TechSupport_No'
    ]

    plt.figure(figsize=(20, 15))
    for i, col in enumerate(selected_categorical_for_churn_viz):
        if col == 'SeniorCitizen' or col == 'gender' or col == 'Partner' or col == 'Dependents' or col == 'PhoneService' or col == 'PaperlessBilling':
             # 对于这些原始二元特征（0/1编码），直接计算平均值即为1类别的流失率
            churn_rate = df_processed.groupby(col)[TARGET_COLUMN].mean().reset_index()
            plt.subplot(4, 3, i + 1)
            sns.barplot(x=col, y=TARGET_COLUMN, data=churn_rate, palette='plasma')
            plt.title(f'{col} 对流失率的影响')
            plt.ylabel('流失率')
            plt.xlabel('')
            plt.xticks(ticks=[0, 1], labels=['否/男性/非老年', '是/女性/老年']) # 示例标签，需要根据具体特征调整
        elif col in df_processed.columns and df_processed[col].dtype == 'int64': # 确保是0/1编码的独热列
            churn_rate = df_processed.groupby(col)[TARGET_COLUMN].mean().reset_index()
            plt.subplot(4, 3, i + 1)
            sns.barplot(x=col, y=TARGET_COLUMN, data=churn_rate, palette='plasma')
            plt.title(f'{col} 对流失率的影响')
            plt.ylabel('流失率')
            plt.xlabel('')
            plt.xticks(ticks=[0, 1], labels=['否', '是'])
        else:
            # 针对原始多分类特征（需要在预处理前处理，或通过独热编码后的列组合分析）
            # 这里我们假设已经独热编码，直接使用独热编码后的列
            pass # 已经通过 selected_categorical_for_churn_viz 处理了

    plt.tight_layout()
    plt.savefig(os.path.join(FIGURES_DIR, 'categorical_churn_rates.png'))
    plt.show()

    # 流失率与数值特征的关系
    print("\n--- 流失率与数值特征的关系 ---")
    plt.figure(figsize=(15, 6))
    #使用 histplot 并通过 hue=TARGET_COLUMN 参数，将数值特征的分布按是否流失进行区分，从而直观地看到流失客户和非流失客户在这些特征上的差异。
    # Tenure vs Churn
    plt.subplot(1, 3, 1)
    sns.histplot(data=df_processed, x='tenure', hue=TARGET_COLUMN, kde=True, palette='viridis', stat='density', common_norm=False)
    plt.title('Tenure (客户时长) 与流失')
    plt.xlabel('客户时长 (标准化)')
    plt.ylabel('密度')

    # MonthlyCharges vs Churn
    plt.subplot(1, 3, 2)
    sns.histplot(data=df_processed, x='MonthlyCharges', hue=TARGET_COLUMN, kde=True, palette='viridis', stat='density', common_norm=False)
    plt.title('MonthlyCharges (月费用) 与流失')
    plt.xlabel('月费用 (标准化)')
    plt.ylabel('密度')

    # TotalCharges vs Churn
    plt.subplot(1, 3, 3)
    sns.histplot(data=df_processed, x='TotalCharges', hue=TARGET_COLUMN, kde=True, palette='viridis', stat='density', common_norm=False)
    plt.title('TotalCharges (总费用) 与流失')
    plt.xlabel('总费用 (标准化)')
    plt.ylabel('密度')

    plt.tight_layout()
    plt.savefig(os.path.join(FIGURES_DIR, 'numerical_churn_relationship.png'))
    plt.show()

    # 特定群体分析：客户类型细分（例如：无互联网服务客户的流失情况）
    # 假设我们有一个 'InternetService_No' 列 (独热编码后)
    if 'InternetService_No' in df_processed.columns:
        print("\n--- 无互联网服务客户的流失情况 ---")
        no_internet_churn = df_processed[df_processed['InternetService_No'] == 1][TARGET_COLUMN].value_counts(normalize=True) * 100
        print(f"无互联网服务客户流失率:\n{no_internet_churn}")

        plt.figure(figsize=(6, 4))
        sns.barplot(x=no_internet_churn.index, y=no_internet_churn.values, palette='viridis')
        plt.title('无互联网服务客户流失分布')
        plt.xlabel('流失')
        plt.ylabel('百分比 (%)')
        plt.xticks(ticks=[0, 1], labels=['不流失', '流失'])
        plt.savefig(os.path.join(FIGURES_DIR, 'no_internet_churn_distribution.png'))
        plt.show()


else:
    print("预处理数据加载失败，请检查路径和文件是否存在。")

print("\n更深入的数据探索完成。")
```
